```
'''
HIGH-PASS Filter
--> Sharpen an image
--> Enhance high-frequency parts of an image
'''
```

```
import cv2 # computer vision library
#import helpers

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```
image = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Color Detection/Data File/254554.jpg')
```

```
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```
# 3x3 array for edge detection
#Sobel filter
sobel_x = np.array([
                   [-1,0,1],
                   [-2,0,2],
                   [-1,0,1]
])
#Perform convulation using filter2D
filter_img = cv2.filter2D(gray,-1,sobel_x)
plt.imshow(filter_img,cmap='gray')
```

```
sobel_x = np.array([
                   [1,0,-1],
                   [2,0,-2],
                   [1,0,-1]
])
#Perform convulation using filter2D
filter_img = cv2.filter2D(gray,-1,sobel_x)
plt.imshow(filter_img,cmap='gray')
```

```
sobel_x = np.array([
                   [0,1,-1],
                   [0,2,-2],
                   [0,1,-1]
])
#Perform convulation using filter2D
filter_img = cv2.filter2D(gray,-1,sobel_x)
plt.imshow(filter_img,cmap='gray')
```

```
sobel_x = np.array([
                   [-1,0,1],
                   [-2,0,2],
                   [-1,0,1]
])
#Perform convulation using filter2D
filter_img = cv2.filter2D(gray,-1,sobel_x)
plt.imshow(filter_img,cmap='gray')
```

```
# Create threshold that sets all the filtered pixels to white
# Above a certain thershold
r, binary_img = cv2.threshold(filter_img, 20, 255, cv2.THRESH_BINARY)
plt.imshow(binary_img, cmap='gray')
```
