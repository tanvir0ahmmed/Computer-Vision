```
import cv2 # computer vision library
#import helpers

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```
night_image = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/DayNight Classifier /vincent-chin-KRVmB7wraZ4-unsplash.jpg')
day_image = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/DayNight Classifier /gustavo-quepon-pF_2lrjWiJE-unsplash.jpg')
```

```
n_img = np.copy(night_image)
d_img = np.copy(day_image)
```

```
n_img = cv2.cvtColor(n_img, cv2.COLOR_BGR2RGB)
plt.imshow(n_img)
```

```
d_img = cv2.cvtColor(d_img, cv2.COLOR_BGR2RGB)
plt.imshow(d_img)
```

```
#Plot Color Channels

#RGB channels

r = n_img[:,:,0]
g = n_img[:,:,1]
b = n_img[:,:,2]

f, (ax1,ax2,ax3) =  plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
```

```
#Plot Color Channels

#RGB channels

r = d_img[:,:,0]
g = d_img[:,:,1]
b = d_img[:,:,2]

f, (ax1,ax2,ax3) =  plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
```

```
#HSV channels
n_hsv = cv2.cvtColor(n_img, cv2.COLOR_RGB2HSV)
h = n_hsv[:,:,0]
s = n_hsv[:,:,1]
v = n_hsv[:,:,2]

f, (ax1,ax2,ax3) =  plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```
#HSV channels
d_hsv = cv2.cvtColor(d_img, cv2.COLOR_RGB2HSV)
h = d_hsv[:,:,0]
s = d_hsv[:,:,1]
v = d_hsv[:,:,2]

f, (ax1,ax2,ax3) =  plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```
sum_brightness = np.sum(d_hsv[:,:,2])
area = d_hsv.shape[0]*d_hsv.shape[1]  # pixels
d_avg = sum_brightness/area
print(d_avg)

sum_brightness = np.sum(n_hsv[:,:,2])
area = n_hsv.shape[0]*n_hsv.shape[1]  # pixels
n_avg = sum_brightness/area
print(n_avg)
```
