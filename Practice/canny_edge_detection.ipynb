```
'''
Canny Edge Detection
--> Eliminates weak edges and noise
--> Isolates edges that are part of an object boundary 

Algorithm
1. Filters out noise using a Gausssian blur
2. Finds the strength and direction of edges using Sobel filters
3. Applies non-maximum suppression to issolate the strogest edges and thin to one-pixel wide lines
4. Uses hysteresis(double threshold) to isolate the best edges
'''
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```
image = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Color Detection/Data File/735920.jpg')
```

```
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap='gray')
```

```
# Define lower and upper thresholds for hysteresis
lower = 120
upper = 240
edges = cv2.Canny(gray,lower,upper)
plt.imshow(edges, cmap='gray')
```

```
# using 'wide' & 'tight' threshold
wide = cv2.Canny(gray, 30, 100)
tight = cv2.Canny(gray, 180, 240)

f,(ax1,ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.set_title('wide')
ax1.imshow(wide, cmap='gray')

ax2.set_title('tight')
ax2.imshow(tight, cmap='gray')
```
