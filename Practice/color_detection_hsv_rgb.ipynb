```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Color Detection/Data File/440969.jpg')
print(type(image))
print(image.shape)
```

```
%matplotlib inline

img = np.copy(image)
# OpenCv reads in color image as BGR, thats why use BGR2RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```
#Plot Color Channels

#RGB channels

r = img[:,:,0]
g = img[:,:,1]
b = img[:,:,2]

f, (ax1,ax2,ax3) =  plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Red')
ax1.imshow(r, cmap='gray')

ax2.set_title('Green')
ax2.imshow(g, cmap='gray')

ax3.set_title('Blue')
ax3.imshow(b, cmap='gray')
```

```
#HSV channels
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

f, (ax1,ax2,ax3) =  plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Hue')
ax1.imshow(h, cmap='gray')

ax2.set_title('Saturation')
ax2.imshow(s, cmap='gray')

ax3.set_title('Value')
ax3.imshow(v, cmap='gray')
```

```
lower_red = np.array([120,0,0])
upper_red = np.array([208,96,110])

lower_hue = np.array([120,80,90])
upper_hue = np.array([255,170,210])
```

```
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)
mask_img = np.copy(img)
mask_img[mask_hsv==0] = [0,0,0]
plt.imshow(mask_img)
plt.title('HSV SELECTION')
```

```
mask_rgb = cv2.inRange(img, lower_red, upper_red)
mask_img = np.copy(img)
mask_img[mask_rgb==0] = [0,0,0]
plt.imshow(mask_img)
plt.title('RGB SELECTION')
```
