```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
image = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Color Detection/Data File/254554.jpg')
print(type(image))
print(image.shape)
```

```
image = image[0:2000, 0:3200]
print(image.shape)
```

```
%matplotlib inline

img = np.copy(image)
# OpenCv reads in color image as BGR, thats why use BGR2RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```
 plt.imshow(img)
```

```
# define bounds for isolate the color that i want(background green color)
lower_range = np.array([0, 176, 0])
upper_range = np.array([120, 196, 62])

#Masks are a very common way to isolate a selected area of interest and do something with that area
'''
inRange function takes an image in lower and upper bounds andf defines a mask by asking if 
the color value of each image pixel falls in the range of the lower and upper color threshilds.
if it does fall in this range, the mask will allow it to be displayed and if not it will block it out and turn the pixel black
'''
mask = cv2.inRange(img, lower_range, upper_range)
print(mask)
plt.imshow(mask, cmap='gray')
```

```
mask_img = np.copy(img)
mask_img[mask != 0] = [0,0,0]
plt.imshow(mask_img)
```

```
background = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Color Detection/Data File/1098239.jpg')
print(background.shape)
background = cv2.cvtColor(background, cv2.COLOR_BGR2RGB)
crop_background = background[0:2000, 0:3200]
crop_background[mask == 0] = [0,0,0]
plt.imshow(crop_background)
```

```
complete_img = crop_background + mask_img
plt.imshow(complete_img)
```
