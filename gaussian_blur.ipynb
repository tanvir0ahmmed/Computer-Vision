```
'''
Gaussian Blur
--> Blur/Smooth and image
--> Block high-frequency parts of an image
--> Preserve edges!
'''
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

%matplotlib inline
```

```
imageA = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Edge Detection/IMG_20201024_204928.jpg')
imageB = cv2.imread('/content/drive/My Drive/UpWork/Computer Vision/Practice/Edge Detection/pierre-bamin-_EzTds6Fo44-unsplash.jpg')
```

```
imageA = cv2.cvtColor(imageA, cv2.COLOR_BGR2RGB)
plt.imshow(imageA)
```

```
imageB = cv2.cvtColor(imageB, cv2.COLOR_BGR2RGB)
plt.imshow(imageB)
```

```
# Convert to grayscale for filtering 
grayA = cv2.cvtColor(imageA, cv2.COLOR_RGB2GRAY)
grayB = cv2.cvtColor(imageB, cv2.COLOR_RGB2GRAY)
```

```
#Create a Gaussian blurred image
gray_blurA = cv2.GaussianBlur(grayA, (5,5), 0)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(grayA, cmap='gray')

ax2.set_title('blurred image')
ax2.imshow(gray_blurA, cmap='gray')
```

```
gray_blurB = cv2.GaussianBlur(grayB, (5,5), 0)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(grayB, cmap='gray')

ax2.set_title('blurred image')
ax2.imshow(gray_blurB, cmap='gray')
```

```
#Test with a high-pass filter

sobel = np.array([
                  [-2,0,2],
                  [-1,0,1],
                  [-2,0,2]
])

filterd = cv2.filter2D(grayA,-1,sobel)
filtered_blured = cv2.filter2D(gray_blurA,-1,sobel)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(filterd, cmap='gray')

ax2.set_title('blurred image')
ax2.imshow(filtered_blured, cmap='gray')
```

```
sobel = np.array([
                  [-2,0,2],
                  [-1,0,1],
                  [-2,0,2]
])

filterd = cv2.filter2D(grayB,-1,sobel)
filtered_blured = cv2.filter2D(gray_blurB,-1,sobel)

f, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,10))

ax1.set_title('original gray')
ax1.imshow(filterd, cmap='gray')

ax2.set_title('blurred image')
ax2.imshow(filtered_blured, cmap='gray')
```
